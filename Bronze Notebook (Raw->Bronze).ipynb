```python
df_orders=spark.read.format('csv').option("header","true")\
    .load('/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_orders_dataset.csv')

df_products=spark.read.format('csv').option("header","true")\
    .load('/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_products_dataset.csv')

df_order_items=spark.read.format('csv').option("header","true")\
    .load('/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_order_items_dataset.csv')

df_customers=spark.read.format('csv').option("header","true")\
    .load('/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_customers_dataset.csv')

df_payments=spark.read.format('csv').option("header","true")\
    .load('/Volumes/main/ecommerce/lakehouse_vol/Raw_DS/olist_order_payments_dataset.csv')
```
